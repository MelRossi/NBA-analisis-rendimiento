### “Rendimiento Mínimo Óptimo” (RMO)

Umbral estadístico que define el nivel mínimo de desempeño aceptable o “óptimo” en una métrica determinada (puntos, asistencias, eficiencia, etc.) según los datos históricos reales.

Aplicación en all_seasons

```python
# Calcular RMO por métrica
metricas = ['pts', 'ast', 'reb', 'net_rating']

rmo_values = {}
for m in metricas:
    rmo_values[m] = all_seasons[m].quantile(0.75)

rmo_df = pd.DataFrame.from_dict(rmo_values, orient='index', columns=['RMO'])
rmo_df
```

### Clasificar jugadores según su rendimiento vs. RMO

```python
def classify_rmo(row):
    if row['pts'] >= rmo_values['pts'] and \
       row['net_rating'] >= rmo_values['net_rating']:
        return 'Alto rendimiento'
    elif row['net_rating'] >= (rmo_values['net_rating'] * 0.85):
        return 'Rendimiento aceptable'
    else:
        return 'Bajo rendimiento'

all_seasons['RMO_status'] = all_seasons.apply(classify_rmo, axis=1)

all_seasons['RMO_status'].value_counts()
```

```python
# Graficamos
plt.figure(figsize=(10,5))
sns.kdeplot(data=all_seasons, x='net_rating', fill=True, color='#F24405')
plt.axvline(rmo_values['net_rating'], color='red', linestyle='--', label=f"RMO ({rmo_values['net_rating']:.2f})")
plt.title('Distribución de Eficiencia con Umbral RMO')
plt.xlabel('Eficiencia')
plt.legend()
plt.show()
```

### Interpretación:

- El eje X muestra los valores de eficiencia de todos los jugadores (o registros).

- La curva naranja representa la distribución (qué tan frecuentes son los valores).

- La línea roja vertical marca el punto de corte donde comienza el rendimiento óptimo (superior al RMO).

### Conclusiones:

- La mayoría de los jugadores están concentrados cerca de 0 o valores bajos, es decir, la curva es muy alta y estrecha en el centro → indica que la gran mayoría tiene eficiencia promedio o baja.

- El umbral RMO = 3,2 está muy cerca del pico → eso significa que pocos jugadores superan ese nivel, es decir, el rendimiento alto es poco frecuente.

- La cola derecha (valores positivos más altos) corresponde al grupo élite de rendimiento, donde están los jugadores más destacados.

```python
# Filtrar solo los jugadores con alto rendimiento
rmo_alto = all_seasons[all_seasons['RMO_status'] == 'Alto rendimiento']

# Agrupar para obtener promedio por jugador
top_rmo = (
    rmo_alto.groupby('player_id')[['pts', 'ast', 'reb', 'net_rating']]
    .mean()
    .sort_values('net_rating', ascending=False)
    .head(10)
    .reset_index()
)

# Traer los nombres de los jugadores
top_rmo = top_rmo.merge(
    player[['id', 'full_name']],
    left_on='player_id',
    right_on='id',
    how='left'
)

# Graficar
plt.figure(figsize=(10,5))
sns.barplot(
    y='full_name',
    x='net_rating',
    data=top_rmo,
    color='#F24405'
)
plt.title('Top 10 jugadores con rendimiento superior al RMO')
plt.xlabel('Eficiencia promedio (Net Rating)')
plt.ylabel('Jugador')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# IDs de top 10 global vs RMO
top_jugadores_ids = top_jugadores['player_id'].tolist()
top_rmo_ids = top_rmo['player_id'].tolist()

interseccion = set(top_jugadores_ids) & set(top_rmo_ids)
coincidencia = len(interseccion) / 10 * 100

print(f"Coincidencia entre Top 10 global y Top 10 RMO: {coincidencia:.1f}%")
print("Jugadores en común:", [player.loc[player['id'].isin(interseccion), 'full_name'].tolist()])
```

### Qué significa que no haya coincidencia (0%) con el Top 10 global

El Top 10 global anterior medía el promedio general de eficiencia (net_rating) sin distinguir entre temporadas, volumen de partidos o estabilidad en el tiempo.

En cambio, el Top 10 RMO identifica a los jugadores que:

Superaron el umbral estadístico de RMO (3.2 en eficiencia), y que además, mantienen ese rendimiento de manera consistente (en promedio por jugador).

Jugadores como RJ Hunter, Mac McClung o Ndudi Ebi no son históricamente “superestrellas”,
pero sus registros en el dataset muestran valores altos en Net Rating en pocas apariciones o temporadas, lo que estadísticamente los ubica por encima del umbral RMO.

-----------------------------------------------------------------------------------------------------------------------------------------------------------
Al comparar el Top 10 de eficiencia global con el Top 10 de jugadores que superan el umbral RMO, observamos una ausencia total de coincidencia (0%).
Esto indica que el RMO no necesariamente identifica a los jugadores más reconocidos o constantes, sino a aquellos que tuvieron un rendimiento estadísticamente excepcional en determinados momentos o contextos.

En otras palabras, mientras el Top 10 global refleja el talento sostenido, el Top 10 RMO permite detectar potencial emergente o rendimiento explosivo puntual, útil para scouting y análisis de talento oculto.


El RMO funciona como una métrica complementaria al rendimiento global:
identifica outliers positivos y potenciales talentos no consolidados, mientras que el Top 10 global refleja rendimientos sostenidos en el tiempo.

La diferencia entre ambos rankings demuestra que el modelo de RMO aporta valor para el scouting y la detección de talento emergente, más allá del desempeño promedio histórico.
