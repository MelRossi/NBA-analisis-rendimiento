### Tratamiento de valores nulos en la tabla `line_score`

####  Contexto
La tabla `line_score` contiene estadsticas por partido, incluyendo puntos por cuarto y por tiempos extra (OT).  
Al revisar los datos, se observa que varias columnas presentan valores nulos (`NaN`), especialmente en los tiempos extra superiores (OT5 a OT10).  
Esto ocurre porque la mayora de partidos **no llegan a tantos tiempos extra**, por lo que no existen registros en esas columnas.

---

####  Estrategia de imputacin

#### 1. Columnas de OT (`pts_otX_home`, `pts_otX_away`)
- **Significado del nulo:** el partido no tuvo ese tiempo extra.
- **Accin:** reemplazar por `0`, ya que el equipo no anot en un perodo inexistente.

```python
## Limpieza de columnas de overtime (`pts_ot`)

# Paso 1: Filtrar columnas que contienen "pts_ot"
ot_cols = [col for col in df_line_score.columns if "pts_ot" in col]

# Paso 2: Rellenar valores nulos con 0 en esas columnas
df_line_score[ot_cols] = df_line_score[ot_cols].fillna(0)
df_line_score[ot_cols].isnull().sum()  ## Verificacin de valores nulos en las columnas de overtime tras la limpieza
```

```python
# Duplicados exactos
duplicados = df_line_score[df_line_score.duplicated()]
print("Duplicados exactos:", duplicados.shape)

# Duplicados por claves de partido y equipo
duplicados_claves = df_line_score[df_line_score.duplicated(subset=["game_id","team_id_home","team_id_away"])]
print("Duplicados por claves:", duplicados_claves.shape)
```

```python
# Mostrar todas las filas duplicadas exactas
duplicados = df_line_score[df_line_score.duplicated(keep=False)]
print(duplicados)
```

```python
duplicados_claves = df_line_score[df_line_score.duplicated(subset=["game_id","team_id_home","team_id_away"], keep=False)]
print(duplicados_claves)
```

```python
df_line_score = df_line_score.drop_duplicates(subset=["game_id","team_id_home","team_id_away"], keep="first")
print("Nueva forma tras eliminar duplicados:", df_line_score.shape)
```

### Una vez retirados las filas duplicadas, procedemos a Normalizacin (formato largo)

#### Convierte la tabla a una estructura con una fila por equipo en cada partido.

```python
# Crear tabla para equipos locales
home_cols = [
    "game_date_est","game_id","team_id_home","team_abbreviation_home",
    "team_city_name_home","team_nickname_home","team_wins_losses_home",
    "pts_qtr1_home","pts_qtr2_home","pts_qtr3_home","pts_qtr4_home",
    "pts_ot1_home","pts_ot2_home","pts_ot3_home","pts_ot4_home",
    "pts_ot5_home","pts_ot6_home","pts_ot7_home","pts_ot8_home",
    "pts_ot9_home","pts_ot10_home","pts_home"
]

df_home_line_score = df_line_score[home_cols].copy()
df_home_line_score = df_home_line_score.rename(columns=lambda x: x.replace("_home",""))
df_home_line_score["local_visitante"] = "home"
print(df_home_line_score.head())
```

```python
# Crear tabla para equipos visitantes
away_cols = [
    "game_date_est","game_id","team_id_away","team_abbreviation_away",
    "team_city_name_away","team_nickname_away","team_wins_losses_away",
    "pts_qtr1_away","pts_qtr2_away","pts_qtr3_away","pts_qtr4_away",
    "pts_ot1_away","pts_ot2_away","pts_ot3_away","pts_ot4_away",
    "pts_ot5_away","pts_ot6_away","pts_ot7_away","pts_ot8_away",
    "pts_ot9_away","pts_ot10_away","pts_away"
]

df_away_line_score = df_line_score[away_cols].copy()
df_away_line_score = df_away_line_score.rename(columns=lambda x: x.replace("_away",""))
df_away_line_score["local_visitante"] = "away"
print(df_away_line_score.head())

```

```python
# Unir ambas tablas
df_line_score_filtrado = pd.concat([df_home_line_score, df_away_line_score], ignore_index=True)

# Verificar resultado
print(df_line_score_filtrado.head())
print(df_line_score_filtrado.shape)
```

# Normalizacin de la tabla `line_score` (Formato largo)

## Objetivo
La tabla `line_score` originalmente est en **formato ancho**, con columnas separadas para equipos locales (`home`) y visitantes (`away`).  
Este formato dificulta el anlisis porque cada partido est representado en una sola fila con dos equipos.  
La normalizacin busca transformar la tabla a **formato largo**, donde cada fila corresponde a **un equipo en un partido**.

---

## Estrategia
1. **Separar equipos locales y visitantes** en dos DataFrames (`df_home` y `df_away`).  
2. **Renombrar columnas** para eliminar el sufijo `_home` y `_away`.  
3. **Agregar columna `local_visitante`** para indicar si el equipo jug como local o visitante.  
4. **Unir ambos DataFrames** en uno solo (`df_line-.score_filtrado`), duplicando el nmero de filas (dos por partido).
